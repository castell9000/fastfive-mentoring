일단 숫자로만 이루어진 영수증 정보를 적당히 알아 볼 수 있게 만들었어요.

영수증정보와 고객, 제품, 직원 정보를 합치는건 계속 쓸것 같으니까 가볍게 함수로 만들고

좀 더 보기 편하게 정제하는 함수를 하나 짜보도록 하죠

```python
def mergeBasicReceiptInfo(receipt, product, customer, staff):
    # 각각 하나씩 데이터를 합쳐줍시다
    # 추가로 필요한 정보만 들고 와줍시다
    tempFirst = pd.merge(receipt[['transaction_id', 'transaction_date', 'transaction_time', 'staff_id', 'customer_id','product_id', 'quantity', 'unit_price']],
                         product[['product_id', 'product_type', 'product', 'current_retail_price']], on = 'product_id')
    tempSecond = pd.merge(tempFirst,
                         customer[['customer_id', 'customer_first-name']], on = 'customer_id')
    tempThird = pd.merge(tempSecond,
                        staff[['staff_id', 'first_name', 'last_name']], on = 'staff_id')
    
    # 합치거나 곱해서 보면 좋은 정보는 합치고 곱해줍시다
    #스태프 이름
    tempThird['first_name'] = tempThird['first_name'] + ' '+ tempThird['last_name']
    #거래시간
    tempThird['transaction_date'] = tempThird['transaction_date'] + ' ' +tempThird['transaction_time']
    #총합 금액
    tempThird['sum_price'] = tempThird['unit_price'] * tempThird['quantity']
    mergeColumn = tempThird

    # 보기 힘든 이름은 조금더 이해하기 쉬운 이름으로
    renameColumn = mergeColumn.rename(columns={'first_name' : 'staff_name','customer_first-name' : 'customer_name',
                                               'unit_price' : 'price', 'current_retail_price' : 'original_price'})
    # 데이터를 합쳤기에 생기는 중복 컬럼 없애주기
    dropColumn = renameColumn.drop(['transaction_time', 'last_name'], axis=1)
    # 보기 편하게 컬럼의 순서를 바꿔줍시다
    sortColumn = dropColumn[['transaction_id', 'transaction_date', 'product', 'product_type','sum_price',
                             'customer_name', 'staff_name','price', 'quantity',
                            'product_id', 'customer_id', 'staff_id']]
    
    # 돌아간 결과를 transaction id 기준으로 다시 정렬해줍시다
    sortColumn = sortColumn.sort_values(by=['transaction_id'], axis=0).reset_index(drop=True)
    # 결과를 반환해줍시다
    return sortColumn
```

```python
# 영수증정보, 제품정보, 고객정보, 직원정보 순으로 함수에 넣어서 결과를 받아봅시다.
mergeBasicReceiptInfo(receiptNo1Clean, product, customer, staff)
```

```python
# 모든 영수증 데이터
mergeBasicReceiptInfo(recieptsIndex, product, customer, staff)

# 영수증데이터는 분명 49894 건 이였는데, 합치고나니까 좀 줄어들었네요..?
# 그럼 여기서 우리는 데이터에 결측치가 좀 있다는것을 알 수 있습니다.
# 나중에 또 결측치를 찾아서 처리해줘야 하겠지만.. 이까지만 해도 충분히 많이 했죠?
```
